```python
loan_amount = ct_data["loan_amount_000s"].values
loan_amount = loan_amount.reshape(-1, 1)
app_income = ct_data["applicant_income_000s"].values
app_income = app_income.reshape(-1,1)
location = ct_data["county_name"].values
location = location.reshape(-1,1)
action = ct_data["action_taken"].values
```

```python
loan_amount_train, loan_amount_test, action_train, action_test = train_test_split(loan_amount, action, train_size=0.8, test_size=0.2, random_state=100)
app_income_train, app_income_test, actionapp_income_train, actionapp_income_test = train_test_split(app_income, action, train_size=0.8, test_size=0.2, random_state=100)

lm1 = LinearRegression()
lm1.fit(loan_amount_train,action_train)
loan_amount_action_predict = lm1.predict(loan_amount_test)

lm2 = LinearRegression()
lm2.fit(app_income_train,actionapp_income_train)
app_incomeaction_predict = lm2.predict(app_income_test)
```
